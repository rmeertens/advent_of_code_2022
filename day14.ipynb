```python
from collections import defaultdict
lines = open('inputs/day14.txt').readlines()
lines = [line.strip() for line in lines]

def get_cave(has_floor = False): 
    cave = defaultdict(lambda: '.')
    max_y = 0
    for line in lines: 
        coordinates = line.split('->')
        
        for start in range(len(coordinates)-1):
            start_str = coordinates[start]
            end_str = coordinates[start+1]
            
            x_start, y_start = start_str.split(',')
            x_end, y_end = end_str.split(',')
            x_start, y_start = int(x_start.strip()), int(y_start.strip())
            x_end, y_end = int(x_end.strip()), int(y_end.strip())
            
            x_start, x_end = min(x_start, x_end), max(x_start, x_end)
            y_start, y_end = min(y_start, y_end), max(y_start, y_end)
            
            max_y = max(max_y, y_end)
            
            for x in range(x_start, x_end+1):
                for y in range(y_start, y_end+1):
                    cave[(x,y)] = '#'
    if has_floor:
        floor_level = max_y + 2
        # Create the bottom of the cave
        for x in range(500 - 2* floor_level, 500 + 2* floor_level):
            cave[(x, floor_level)] = '#'
    return cave
    
def simulate_particles(cave):     
    max_y = max([y for x,y in cave.keys()])
    source_sand = (500, 0) # X, Y

    came_to_rest = True
    particles_placed = 0
    while cave[source_sand] == '.' and came_to_rest:
        # Simulate particles falling from source till they hit another particle or the bottom

        new_particle = source_sand 

        still_falling = True
        while still_falling:
            possible_places = [(new_particle[0], new_particle[1]+1), # Down 
                                (new_particle[0]-1, new_particle[1]+1), # Down left
                                (new_particle[0]+1, new_particle[1]+1 ) # Down right
                                ]
            still_falling = False
            for place in possible_places:
                if cave[place] == '.':
                    new_particle = place
                    still_falling = True
                    break

            if new_particle[1] > max_y:
                still_falling = False
                came_to_rest = False
                break

        cave[new_particle] = 'o'
        particles_placed += came_to_rest
    return particles_placed



cave = get_cave(has_floor=False)
particles_placed = simulate_particles(cave)
print('Part 1', particles_placed)

cave = get_cave(has_floor=True)
particles_placed = simulate_particles(cave)
print('Part 2', particles_placed)
```
