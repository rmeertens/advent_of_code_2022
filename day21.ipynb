```python
import re
lines = open('inputs/day21.txt').readlines()
lines = [line.strip() for line in lines]

def get_root(lines):
    monkey_numbers = dict()
    while not "root" in monkey_numbers:
        for line in lines:
            monkey_name, equation = line.split(": ")
            #print(monkey_name, equation)
            try: 
                # evaluate the equation
                monkey_numbers[monkey_name] = eval(equation)
            except:
                pass

            # get all the strings from the equation
            strings = re.findall(r'[\w]+', equation)

            # check if all the strings are in the dictionary
            if all([string in monkey_numbers for string in strings]):
                # replace the strings with their values
                for string in strings:
                    equation = equation.replace(string, str(monkey_numbers[string]))
                monkey_numbers[monkey_name] = eval(equation)

                if monkey_name == "root":
                    return [monkey_numbers[string] for string in strings], monkey_numbers[monkey_name]
    return None

answer = get_root(lines)    
print("Part 1:", answer[1])

# Re-implemented this part such that it uses binary search
low = 0
high = 1e20
while low < high:
    i = (low + high) // 2

    # Replace the line which starts with "humn" with the new value
    lines = [line if not line.startswith("humn") else "humn: " + str(i) for line in lines]
    (num1, num2), answer = get_root(lines)    
    print("Try number", i, "delta", "result", num1, num2)
    
    if num1 == num2:
        print("FOUND! Part 2:" , i)
        break
    elif num1 > num2:
        print("TOO LOW", i)
        low = i
    else: 
        print("TOO HIGH", i)
        high = i
        
        
```

```python
lines = open('inputs/day21.txt').readlines()
lines = [line.strip() for line in lines]

# Build a dictionary mapping names of monkeys to equations they are shouting
name_equation_dict = dict()
for line in lines: 
    name, equation = line.split(": ")
    name_equation_dict[name] = equation


def get_answer(node_name): 
    """
    Recursively evaluates the answer to a question. 
    """
    equation = name_equation_dict[node_name].split(" ")
    if len(equation) == 1:
        return int(equation[0])
    else:
        assert len(equation) == 3
        left = get_answer(equation[0])
        right = get_answer(equation[2])
        
        if node_name == "root":
            return left, right
        else: 
            string_equation = name_equation_dict[node_name]
            string_equation = string_equation.replace(equation[0], str(left))
            string_equation = string_equation.replace(equation[2], str(right))
            return eval(string_equation)

# Use binary search to find the correct answer to part 2
low = 0
high = 1e20
while low < high:
    mid = (low+high)//2
    name_equation_dict["humn"] = str(int(mid))

    a, b = get_answer("root")
    
    if a == b:
        print("Found", mid)
        break
    elif a > b: 
        low = mid+1
    else:
        high = mid
```
