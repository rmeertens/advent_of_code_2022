```python
from collections import defaultdict

lines = open('inputs/day07.txt').readlines()
lines = [line.strip() for line in lines]

computer_layout = {
    '': dict()
    }
current_directory = '/'
for line in lines[1:]:
    if line.startswith('$ ls'):
        # ignore the lines with ls
        continue
    elif line.startswith('$ cd'):
        # update folder if starts with cd
        new_path = line.split(' ')[-1]
        if new_path == '..':
            print('moving up to a parent directory', current_directory)
            folders = current_directory.split('/')
            to_keep = [f for f in folders if len(f) > 0][:-1]
            print(folders, 'keeping', to_keep)
            if len(to_keep) == 0:
                current_directory = '/'
            else:
                current_directory = '/' + '/'.join(to_keep) + '/'
            print('moved to parent directory', current_directory)
        else:
            current_directory = current_directory + new_path + '/'
            print('moved to child directory', current_directory)
        
    else:
        cur_folder = computer_layout
        print('info about', current_directory, current_directory.split())
        for folder in current_directory.split('/')[:-1]:
            cur_folder = cur_folder[folder]
        # If the line is a folder, add it to the current directory
        if line.startswith('dir'):
            folder_name = line.split(' ')[-1]
            if folder_name not in cur_folder:
                cur_folder[folder_name] = dict()
        else: 
            # If the line is a file, add the size and name to the current directory
            filesize, filename = line.split(' ')[0], line.split(' ')[-1]
            cur_folder[filename] = int(filesize)

    
```

```python
"""
To begin, find all of the directories with a total size of at most 100000, 
then calculate the sum of their total sizes. 
In the example above, these directories are a and e; 
the sum of their total sizes is 95437 (94853 + 584). 
(As in this example, this process can count files more than once!)

Find all of the directories with a total size of at most 100000. 
What is the sum of the total sizes of those directories?

"""

total_size_of_small_folders = 0
smallest_size_to_delete = 70000000

space_needed = 8381165 # Example 
space_needed = 4376732 # Actual puzzle
def count_small_folders(folder, threshold):
    global total_size_of_small_folders
    global smallest_size_to_delete
    num_small_folders = 0
    total_size = 0
    for filename, value in folder.items():
        # print(filename, value)
        if isinstance(value, int):
            total_size += value
        else:
            num_small, ts = count_small_folders(value, threshold)
            num_small_folders += num_small
            total_size += ts
    if total_size <= threshold:
        #print('small folder', folder, total_size)
        total_size_of_small_folders += total_size
    if total_size >= space_needed: 
        smallest_size_to_delete = min(smallest_size_to_delete, total_size)
    return num_small_folders + (total_size <= threshold), total_size

threshold = 100000    
num_small_folders, total_space_used = count_small_folders(computer_layout[''], threshold)
print('part 1', total_size_of_small_folders)
print('total space used', total_space_used)

unused_space = 70000000 - total_space_used
space_needed = 30000000 - unused_space
```
